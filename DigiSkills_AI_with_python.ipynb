# **Operators in Python**
**Operators** in Python are special symbols that perform operations on variables and values. They are categorized into several types based on their functionality.
## 1. Arithmetic Operators

**Arithmetic operators** are used to perform mathematical operations such as addition, subtraction, multiplication, and division.

```python
# Addition
print(10 + 5) # Output: 15
# Subtraction
print(10 - 5) # Output: 5
# Multiplication
print(10 * 5) # Output: 50
# Division
print(10 / 5) # Output: 2.0
# Modulus
print(10 % 3) # Output: 1
# Exponentiation
print(2 ** 3) # Output: 8
# Floor Division
print(10 // 3) # Output: 3
```

## 2. Comparison Operators

**Comparison operators** are used to compare two values and return a boolean result (True or False). They include ==, !=, >, <, >=, and <=.

```python
a = 10
b = 20
print(a == b) # Output: False
print(a != b) # Output: True
print(a > b) # Output: False
print(a < b) # Output: True
print(a >= b) # Output: False
print(a <= b) # Output: True
```

## 3. Logical Operators

**Logical operators** are used to combine conditional statements. They include and, or, and not.

```python
print(True and False)
print(True or False)
print(not False)
```

## 4. Assignment Operators

**Assignment operators** are used to assign values to variables. They include operators like =, +=, -=, *=, /=, and more. 

```python
x = 5
x += 3 # Equivalent to x = x + 3
print(x) # Output: 8
x -= 2 # Equivalent to x = x - 2
print(x) # Output: 6
x *= 2 # Equivalent to x = x * 2
print(x) # Output: 12
x /= 3 # Equivalent to x = x / 3
print(x) # Output: 4.0
```

## 5. Identity Operators

**Identity operators** are used to compare the memory locations of two objects. They include **is** and **is not**.

```python
a = 10
b = 10
c = a
print(a is b) # Output: True
print(a is not c) # Output: False
```

## 6. Membership Operators

**Membership operators** are used to test if a sequence is present in an object. They include **in** and **not in**.

```python
x = 'Hello world'
y = {1: 'a', 2: 'b'}
print('H' in x) # Output: True
print('hello' not in x) # Output: True
print(1 in y) # Output: True
print('a' in y) # Output: False
```

## Local and Global Variable

1. A variable declared inside a function is called a **local** variable
2. A variable decalred outside a function is called a **global** variable

```python
def show():
    x = 5
    print(x)
x = 10
show()
```

```python
x = 10
def show():
    global x
    x = 5
    return x
print(show())
```

## **Common use cases:**

1. With **map()** – Apply a transformation to each element:

```python
nums = [1, 2, 3, 4, 5]
doubled = list(map(lambda x: x * 2, nums))
print(doubled) 
```

2. With **filter()** – Select elements matching a condition:

```python
nums = [1, 2, 3, 4, 5, 6, 7, 8, 9]
odds = list(filter(lambda x: x % 2 != 0, nums))
print(odds)
```

3. With **sorted()** – Custom sort:

```python

students = [("Emil", 25), ("Tobias", 22), ("Linus", 28)]
sorted_students = sorted(students, key=lambda x: x[1])
print(sorted_students) # [('Tobias', 22), ('Emil', 25), ('Linus', 28)]
```
